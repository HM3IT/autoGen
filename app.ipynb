```python
# This is older version autogen specifically for AutoGen 2.0

import os
from autogen import ConversableAgent
from dotenv import load_dotenv

load_dotenv()

llm_endpoint = os.environ["OLLAMA_ENDPOINT"]
model_name = os.environ["LLM_MODEL_NAME"]

api_key = os.environ["LLM_API_KEY"]

agent = ConversableAgent(
    "chatbot",
    llm_config={
        "config_list": [
            {
                "model": model_name,                 
                "base_url": llm_endpoint,  # The local Ollama API endpoint
                "api_type": "openai",              
                "api_key": api_key,   
                "price": [0.0, 0.0],                   
            }
        ]
    },
    code_execution_config=False,  # Turn off code execution, by default it is off.
    function_map=None,            # No registered functions, by default it is None.
    human_input_mode="NEVER",     # Never ask for human input.
)



reply = agent.generate_reply(messages=[{"content": "Tell me a joke.", "role": "user"}])
print(reply)
```

```python
cathy = ConversableAgent(
    "cathy",
    system_message="Your name is Cathy and you are a part of a duo of comedians.",
    llm_config={
        "config_list": [{
            "model": model_name,
            "temperature": 0.9,
            "api_key": api_key,
            "base_url": llm_endpoint,     
            "api_type": "openai",   
            "price": [0.0, 0.0],   
        }]
    },
    human_input_mode="NEVER",   
    
)

joe = ConversableAgent(
    "joe",
    system_message="Your name is Joe and you are a part of a duo of comedians.",
    llm_config={
        "config_list": [{
            "model": model_name,
            "temperature": 0.7,
            "api_key": api_key,
            "base_url": llm_endpoint,
            "api_type": "openai",
            "price": [0.0, 0.0],
        }]
    },
    human_input_mode="NEVER",  
    is_termination_msg=lambda msg: "good bye" in msg["content"].lower(),
)
```

```python
result = joe.initiate_chat(cathy, message="Cathy, tell me a joke.", max_turns=3)


## terminating the conversation
# max_turns
# max_consecutive_auto_reply
# is_termination_msg
```

```python
# Human interactions
agent_with_number = ConversableAgent(
    "agent_with_number",
    system_message="You are playing a game of guess-my-number. You have the "
    "number 53 in your mind, and I will try to guess it. "
    "If I guess too high, say 'too high', if I guess too low, say 'too low'. ",
    llm_config={"config_list": [{ 
        "model": model_name,
        "temperature": 0.7,
        "api_key": api_key,
        "base_url": llm_endpoint,
        "api_type": "openai",
        "price": [0.0, 0.0],
    }]},
    is_termination_msg=lambda msg: "53" in msg["content"],  # terminate if the number is guessed by the other agent
    human_input_mode="NEVER",  # never ask for human input
)



human_proxy = ConversableAgent(
    "human_proxy",
    llm_config=False,  # no LLM used for human proxy
    human_input_mode="ALWAYS",  # always ask for human input
)

# Start a chat with the agent with number with an initial guess.
result = human_proxy.initiate_chat(
    agent_with_number,  # this is the same agent with the number as before
    message="10",
)
```
